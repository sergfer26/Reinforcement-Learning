```python
import numpy as np
import pandas as pd
from IPython.core.pylabtools import figsize
from matplotlib import pyplot as plt
from TicTacToe.human import Human as human
from TicTacToe.duel import duel, play_n_duels
```

# Value iteration

## Value iteration player vs random player

```python
from TicTacToe.agent_value_iteration import create_avi
from TicTacToe.agent_random import Random as random_agent
```

```python
agent1 = create_avi()
agent2 = random_agent()
```

```python
playerX = agent2
playerO = agent1
playerO.set_role('O')
```

```python
prob = play_n_duels(3000, playerX, playerO)
```

```python
agent1.wins
```

```python
df = pd.DataFrame({ 'probabilidad' :prob })

figsize(11, 9)
df.plot(title='Juegos ganados por el agente VI ',color='b')
plt.xlabel("10 juegos simulados")
plt.ylabel("Porcentaje de juegos ganados")
```

## Value iteration player vs random min max player

```python
from rdm_minmax_agent import Rdm_MinMax_Agent as rdmMinMax
agent3 = rdmMinMax()
```

```python
agent1.wins = 0
```

```python
playerX = agent3
playerO = agent1
playerO.set_role('O')
```

```python
prob = play_n_duels(3000, playerX, playerO)
```

```python
agent1.wins
```

```python
df = pd.DataFrame({ 'probabilidad' :prob })

figsize(11, 9)
df.plot(title='Juegos ganados por el agente VI ',color='b')
plt.xlabel("10 juegos simulados")
plt.ylabel("Porcentaje de juegos ganados")
```

# Monte Carlo  

## Monte Carlo player vs random player

```python
from TicTacToe.agent_mc import create_amc
```

```python
agent4 = create_amc()
```

```python
playerX = agent2
playerO = agent4
playerX.set_role('X')
playerO.set_role('O')
```

```python
prob = play_n_duels(3000, playerX, playerO)
```

```python
vals = [(k,v) for k, v in agent4.values.items() if v != 0]
```

```python
len(vals)
```

```python
agent4.wins
```

```python
df = pd.DataFrame({ 'probabilidad' :prob })

figsize(11, 9)
df.plot(title='Juegos ganados por el agente MC ',color='r')
plt.xlabel("juegos simulados (10)")
plt.ylabel("Porcentaje de juegos ganados (%)")
```
